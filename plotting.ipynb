```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset
df = pd.read_csv("isthecorellation.csv", delimiter=";")

# Create a plot
plt.figure(figsize=(8, 6))
plt.scatter(df["Year"], df["NL Beer consumption [x1000 hectoliter]"])
plt.xlabel("Year")
plt.ylabel("NL Beer consumption [x1000 hectoliter]")
plt.title("Beer consumption in the Netherlands per year")

# Set the DPI to 300 when saving the plot
plt.savefig("scatter_plot.png", dpi=300)
```
